```python
while True:
    value = input("請輸入整數,輸入(q)離開:")
    if value == "q":
        break    
    value = int(value)
    if value % 2 == 1:
        print(value ** 2)
    else:
        print(value)
print("應用程式結束")
```

```python
#請設計一個程式，讓使用者輸入數值，只有加總正偶數值，不加總正奇數值，如果輸入負數，結束程式。

total = 0
num = 0
while(True):     
    value = int(input(f"請輸入第{num+1}數值:"))   
    if value < 0:
        break
        print("break之後")
    elif value % 2 == 1:
        num += 1
        continue
        print("continue之後")
    else:
        num += 1
        total += value
    
    
print("應用程式結束")
print(f"總共執行{num}")
print(total)
        
```
